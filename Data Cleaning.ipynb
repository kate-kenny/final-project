# Playing with Data

## Importing packages

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
EPA = pd.read_csv("pollution_us_2000_2016.csv")
```

```python
EPA.head()
```

```python
EPA.drop(labels = ["NO2 Units", "O3 Units","SO2 Units", "CO Units"],axis=1)
```

have dates only be years so we can average

```python
EPA["Date Local"] = EPA["Date Local"].map(lambda x: x[0:4])
```

```python
EPA.head()
```

## Averaging data

```python
# # finding unique values
state_codes_unique = EPA["State Code"].unique()
state_codes_unique
```

```python
#how to average i think --> unsure how many of these we want to average
for num in state_codes_unique:
    avg_CO_1st_Max = EPA.loc[EPA["State Code"] == num, "CO 1st Max Value"].mean()
    print(str(num) + "\n" + str(avg_CO_1st_Max))
```
